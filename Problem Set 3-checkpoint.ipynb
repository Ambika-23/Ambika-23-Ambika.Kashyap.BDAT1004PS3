```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
MD = pd.read_csv('us-marriages-divorces-1867-2014.csv')
```

```python
MD
```

```python
MD.plot.line(x="Year", y=["Marriages_per_1000", "Divorces_per_1000"], color = "Green, Red", figsize=(18,9)) 
plt.xlabel('Year', fontsize = 20) 
plt.ylabel('No. of Marriages & Divorces', fontsize = 20) 
plt.title("Comparison between Marriages & Divorces", fontsize=30) 
plt.legend()
plt.show()
```

```python
DF = MD[MD.Year.isin([1900, 1950, 2000])]
DF=DF.drop(columns=["Marriages","Divorces","Population"])
DF = DF.set_index('Year')
DF.plot.bar(figsize=(18,9))
plt.xlabel('Year', fontsize = 20) 
plt.ylabel('Number of Cases', fontsize = 20) 
plt.title("Number of marriages and divorces in U.S. between 1900, 1950, and 2000.", fontsize=30) 
plt.legend() 
plt.show()
```

```python
empData = pd.read_csv("roman-emperor-reigns.csv")
```

```python
empData
```

```python
empData = empData[empData.Cause_of_Death == 'Assassinated']
labels = empData.Emperor 
sizes = empData.Length_of_Reign

fig, ax = plt.subplots(figsize=(20,12))
plt.pie(sizes, autopct='%1.1f%%', startangle=0, textprops={'fontsize': 11})
plt.axis('equal')

plt.title("Assasinated Roman Emperors", fontsize=30)
plt.legend( labels, fontsize = 20)
plt.show()
```
